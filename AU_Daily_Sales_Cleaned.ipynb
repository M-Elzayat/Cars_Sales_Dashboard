```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel("/content/AU_Daily_Sales.xlsx")
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
missing_percentage = df.isnull().mean() * 100
missing_percentage
```

```python
df.drop(columns=["Wind Gust (mph)","Wind Chill (F)"],inplace=True)
```

```python
df.columns
```

```python
df["Temperature (F)"].mean()
```

```python
df["Temperature (F)"] = df["Temperature (F)"].fillna(df["Temperature (F)"].mean())
```

```python
df["Temperature Category"] = df["Temperature Category"].fillna("51-60")
```

```python
df["Humidity (%)"] = df["Humidity (%)"].fillna(df["Humidity (%)"].mean())
```

```python
df["Wind Speed (mph)"] = df["Wind Speed (mph)"].fillna(df["Wind Speed (mph)"].mean())
```

```python
df["Wind Direction"] = df["Wind Direction"].fillna("Unknown")
```

```python
df["Fog (Y/N)"] = df["Fog (Y/N)"].fillna(df["Fog (Y/N)"].mode()[0])
```

```python
df["Rain (Y/N)"] = df["Rain (Y/N)"].fillna(df["Rain (Y/N)"].mode()[0])
```

```python
df["Snow (Y/N)"] = df["Snow (Y/N)"].fillna(df["Snow (Y/N)"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df["Humidity (%)"] = df["Humidity (%)"].astype(int)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["Open Date"] = df["Open Date"].dt.strftime("%#m/%#d/%Y %H:%M:%S")
```

```python
df.dtypes
```

```python
df.drop(columns=["Open Date Formatted"],inplace=True)
```

```python
df["Month Order"] = pd.to_datetime(df["Open Date"]).dt.month
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Days to Make Sale'], bins=30, kde=True, color='royalblue')
plt.title('Distribution of Days to Make Sale')
plt.xlabel('Days to Make Sale')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Temperature Category', y='Temperature (F)', data=df, palette='Set2')
plt.title('Temperature Distribution by Category')
plt.xlabel('Weather Category')
plt.ylabel('Temperature (F)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
fog_rain_snow = df[['Fog (Y/N)', 'Rain (Y/N)', 'Snow (Y/N)']]
for i, col in enumerate(fog_rain_snow.columns):
    plt.subplot(1, 3, i+1)
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f'{col} Count')
    plt.xlabel(col)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
monthly_visibility = df.groupby('Month Order')['Visibility (mi)'].mean().reset_index()
sns.lineplot(data=monthly_visibility, x='Month Order', y='Visibility (mi)', marker='o')
plt.title('Average Visibility by Month')
plt.xlabel('Month')
plt.ylabel('Average Visibility (mi)')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Temperature (F)', 'Humidity (%)', 'Wind Speed (mph)', 'Visibility (mi)']], diag_kind='kde')
plt.suptitle('Pairwise Plot of Weather Variables', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(
    data=df,
    x='Temperature Category',
    y='Days to Make Sale',
    palette='coolwarm',
    errorbar=None
)
plt.title('Average Days to Make Sale by Temperature Category')
plt.xlabel('Temperature Category')
plt.ylabel('Average Days to Make Sale')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
